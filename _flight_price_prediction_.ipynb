```python
X=df1.drop(["Unnamed: 0","price","flight"], axis =1)
y = df1["price"]
```

```python
X
```

```python
df1['class'].value_counts()
```

```python
#Converting Categorical values

X['airline'] = X['airline'].replace({'SpiceJet': 0, 'AirAsia': 1, 'GO_FIRST':2, 'Indigo':3, 'Air_India':4, 'Vistara':5})
X['source_city']  = X['source_city'].replace({'Delhi': 0, 'Mumbai': 1, 'Bangalore':2, 'Kolkata':3, 'Hyderabad':4, 'Chennai':5})
X['destination_city'] = X['destination_city'].replace({'Delhi': 0, 'Mumbai': 1, 'Bangalore':2, 'Kolkata':3, 'Hyderabad':4, 'Chennai':5})
X['departure_time']  = X['departure_time'].replace({'Early_Morning': 0, 'Morning': 1, 'Afternoon':2, 'Evening':3, 'Night':4, 'Late_Night':5})
X['stops'] = X['stops'].replace({'zero': 0, 'one': 1, 'two_or_more':2})
X['arrival_time']  = X['arrival_time'].replace({'Early_Morning': 0, 'Morning': 1, 'Afternoon':2, 'Evening':3, 'Night':4, 'Late_Night':5})
X['class'] = X['class'].replace({'Economy': 0, 'Business': 1})
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, root_mean_squared_error, mean_absolute_error
```

```python
lg = LinearRegression()
lg.fit(X_train,y_train)
y_pred = lg.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
print("r2_score:", r2_score(y_test, y_pred))
```

```python
rmse = root_mean_squared_error(y_test, y_pred)
print("Root Mean Squared Error:", rmse)
```

## **Random Forest**

```python
rfr = RandomForestRegressor()
rfr.fit(X_train,y_train)
y_pred = rfr.predict(X_test)
```

```python
r2 = r2_score(y_test, y_pred)
print("r2_score:", r2)
```

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
rmse = root_mean_squared_error(y_test, y_pred)
print("Root Mean Squared Error:", rmse)
```

```python
ame = mean_absolute_error(y_test, y_pred)
print("mean_absolute_error:", ame)
```

# Price Prediction Model

```python
def predict_Price(airline, source_city, departure_time, stops, arrival_time, destination_city, class_type, duration, days_left):
  features=np.array([[airline, source_city, departure_time, stops, arrival_time, destination_city, class_type, duration, days_left]])
  scaled_features = scaler.transform(features)
  predicted_price = dtr.predict(scaled_features)
  return predicted_price[0]
```

```python
#Prediction
sample_prediction = predict_Price(airline=3, source_city=0, departure_time=0, stops=0, arrival_time=0, destination_city=0, class_type=0, duration=2.17, days_left=10)
print("Predicted Price:", sample_prediction)
```
